```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import sys
from glob import glob
%matplotlib inline
```

### ImageNet - ResNet50 - SGD

```python
df1 = pd.read_csv('train/imagenet/bits_24/maxpool_0.0.csv')
df2 = pd.read_csv('train/imagenet/bits_24/maxpool_1.0.csv')
df3 = pd.read_csv('train/imagenet/bits_24/maxpool_10.0.csv')
df4 = pd.read_csv('train/imagenet/bits_24/maxpool_100.0.csv')
df5 = pd.read_csv('train/imagenet/bits_24/maxpool_1000.0.csv')
df = pd.concat([df1,df2,df3,df4,df5])
df = df.astype({"maxpool": int})
df = df.drop_duplicates(ignore_index=True)
```

```python
sizeFig = 2

fig = plt.figure(num=None, figsize=(sizeFig*1.7, sizeFig), facecolor='w', edgecolor='k')
sns.lineplot(x="epoch", y="accuracy", hue="maxpool", ci= 'sd', 
             data=df, palette=sns.color_palette("hls", len(df['maxpool'].unique())))
plt.xlabel("epoch")#, fontsize='25');
plt.ylabel('Test accuracy')#, fontsize='25');
plt.legend(title = r'$\beta$', ncol=1, bbox_to_anchor=(1, 1))
plt.ylim(bottom=0, top=80)
plt.savefig('plots/24bits/sgd/imagenetResNetTrain24.pdf', bbox_inches='tight')
```
