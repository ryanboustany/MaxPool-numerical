```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import sys
%matplotlib inline
```

### LeNET - MNIST - SGD

```python
df1 = pd.read_csv('train/mnist/lenet/batch_norm_False/sgd/bits_16/results.csv')
df2 = pd.read_csv('train/mnist/lenet/batch_norm_True/sgd/bits_16/results.csv')
df = pd.concat([df1,df2])
df=df[df['epoch'] == 99]
df = df.astype({"maxpool": int})
```

```python
plt.figure(num=None, figsize=(2, 2), facecolor='w', edgecolor='k')
g = sns.catplot(x="maxpool", y="test_accuracy", hue="batch_norm", kind="box", data=df, height=2.5, aspect=2)
g._legend.remove()
plt.xlabel(r'$\beta$', fontsize='12');
plt.ylabel('Test accuracy')#, fontsize='25');
plt.legend(title = "Batch norm", bbox_to_anchor=(1, 1))
plt.ylim(bottom=0.9907, top=0.995)
plt.savefig('plots/16bits/sgd/mnistLeNetTest16.pdf', bbox_inches='tight')
```

```python
df = pd.read_csv('train/mnist/lenet/batch_norm_False/sgd/bits_16/results.csv')
df = df[df['batch_norm'] == False]
df = df.astype({"maxpool": int})
```

```python
sizeFig = 2

fig = plt.figure(num=None, figsize=(sizeFig*1.7, sizeFig), facecolor='w', edgecolor='k')
sns.lineplot(x="epoch", y="train_loss", hue="maxpool", ci= 'sd', 
             data=df, palette=sns.color_palette("hls", len(df['maxpool'].unique())))
plt.xlabel("epoch")#, fontsize='25');
plt.ylabel('Train loss')#, fontsize='25');
plt.legend(title = r'$\beta$', ncol=1, bbox_to_anchor=(1, 1))
plt.ylim(bottom=0, top=0.1)
plt.savefig('plots/16bits/sgd/mnistLeNetTrain16.pdf', bbox_inches='tight')
```
